```python

import numpy as np
from Bio.PDB import PDBParser

# Создаем парсер
parser = PDBParser(QUIET=True)

# Загружаем структуры
# 2RH1 - неактивная, 3SN6 - активная
struct_inactive = parser.get_structure('2RH1', '2rh1.pdb')
struct_active = parser.get_structure('3SN6', '3SN6.pdb')

# Функция для получения координат C-alpha атома
def get_ca_coords(structure, chain_id, res_id):
    # В 2RH1 это цепь A, в 3SN6 - цепь R (рецептор)
    chain = structure[0][chain_id]
    residue = chain[res_id]
    atom = residue['CA'] # Берем альфа-углерод как центр аминокислоты
    return atom.get_coord()

# Номера аминокислот для замера (TM3-TM6 ionic lock)
res_tm3 = 131
res_tm6 = 268

# Координаты для неактивного состояния (2RH1, цепь A)
tm3_inact = get_ca_coords(struct_inactive, 'A', res_tm3)
tm6_inact = get_ca_coords(struct_inactive, 'A', res_tm6)

# Координаты для активного состояния (3SN6, цепь R)
tm3_act = get_ca_coords(struct_active, 'R', res_tm3)
tm6_act = get_ca_coords(struct_active, 'R', res_tm6)

# Считаем расстояние (евклидово)
dist_inactive = np.linalg.norm(tm3_inact - tm6_inact)
dist_active = np.linalg.norm(tm3_act - tm6_act)

print(f"Расстояние TM3-TM6 (неактивное, 2RH1): {dist_inactive:.2f} Å")
print(f"Расстояние TM3-TM6 (активное, 3SN6): {dist_active:.2f} Å")
print(f"Отклонение составило: {dist_active - dist_inactive:.2f} Å")

```
